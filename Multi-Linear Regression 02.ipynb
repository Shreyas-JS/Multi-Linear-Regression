# Model Validation

```python
# Calculating VIF
rsq_r=smf.ols("RandD~Admin+Mrkt_Spend",data=data1).fit().rsquared
vif_r=1/(1-rsq_r)

rsq_r=smf.ols("Admin~RandD+Mrkt_Spend",data=data1).fit().rsquared
vif_a=1/(1-rsq_r)

rsq_r=smf.ols("Mrkt_Spend~Admin+RandD",data=data1).fit().rsquared
vif_m=1/(1-rsq_r)

# Storing VIF values in a data frame
df={'Variables':['RandD','Admin','Mrkt_Spend'],'VIF':[vif_r,vif_a,vif_m]}
Vif_df=pd.DataFrame(df)
Vif_df         # No Collinearity
```

```python
# Residual Analysis
# Test for Normality of Residuals (Q-Q Plot)

import statsmodels.api as sm
qqplot=sm.qqplot(model.resid,line='q') # line = 45 to draw the diagnoal line
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

```python
list(np.where(model.resid<-30000))
```

```python
# Residual Plot for Homoscedasticity
def get_standardized_values( vals ):
    return (vals - vals.mean())/vals.std()
```

```python
plt.scatter(get_standardized_values(model.fittedvalues),
            get_standardized_values(model.resid))

plt.title('Residual Plot')
plt.xlabel('Standardized Fitted values')
plt.ylabel('Standardized residual values')
plt.show()
```

```python
# Residual Vs Regressors
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "RandD", fig=fig)
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "Admin", fig=fig)
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "Mrkt_Spend", fig=fig)
plt.show()
```

```python
# Detecting Influencers/Outliers
# Cooks Distance

model_influence = model.get_influence()
(c, _) = model_influence.cooks_distance
c
```

```python
#Plot the influencers values using stem plot
fig = plt.subplots(figsize=(20, 7))
plt.stem(np.arange(len(data1)), np.round(c, 3))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
#index and value of influencer where c is more than .5
(np.argmax(c),np.max(c))
```

# Improving the model

```python
# Discard the data points which are influencers and reassign the row number (reset_index(drop=True))
data2=data1.drop(data1.index[[49]],axis=0).reset_index(drop=True)
data2
```

```python
while np.max(c)>0.5 :
    model=smf.ols("Profit~RandD+Admin+Mrkt_Spend",data=data2).fit()
    (c,_)=model.get_influence().cooks_distance
    c
    np.argmax(c) , np.max(c)
    data2=data2.drop(data2.index[[np.argmax(c)]],axis=0).reset_index(drop=True)
    data2
else:
    final_model=smf.ols("Profit~RandD+Admin+Mrkt_Spend",data=data2).fit()
    final_model.rsquared , final_model.aic
    print("Thus model accuracy is improved to",final_model.rsquared)
```
